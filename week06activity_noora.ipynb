# 3.a) On average which region has the highest  GDP per capita (current US$)


```python
# Finding the average of GDP based on region
df1 = df[['Region', 'GDP per capita (current US$)']].groupby('Region').mean()
df1
```

```python
# Region with highest GDP
df1.nlargest(1,['GDP per capita (current US$)'])
```

NorthernAmerica has the highest GDP.

# 3.b) On average which region has the highest International trade: Imports (million US$)

```python
# Finding the average of International trade based on region
df2 = df[['Region', 'International trade: Imports (million US$)']].groupby('Region').mean()
df2
```

```python
# Region with highest International trade
df2.nlargest(1,['International trade: Imports (million US$)'])
```

NorthernAmerica has the highest International trade

# 4. Which region has more consistent with respect to GDP per capita (current US$)

```python
# Finding the variance of GDP per capita based on regions
df3 = df[['Region', 'GDP per capita (current US$)']].groupby('Region').var()
df3
```

```python
# Region which is more consistent
df3['GDP per capita (current US$)'].idxmin()
```

Western Africa is the region with least variance. So more consistent region is WesternAfrica.
